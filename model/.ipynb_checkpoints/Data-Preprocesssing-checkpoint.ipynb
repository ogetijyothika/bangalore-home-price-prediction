```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    try:
        if len(tokens) == 2:
                return (float(tokens[0])+float(tokens[1]))/2
            
        elif "Sq. Meter" in x:
            pattern = r'(\d+(\.\d+)?)'
            match = re.search(pattern, x)
            if match:
                number = float(match.group(0))
                return round(number * 10.7639)
            
        elif "Perch" in x:
            pattern = r'(\d+(\.\d+)?)'
            match = re.search(pattern, x)
            if match:
                number = float(match.group(0))
                return round(number * 272.25)
        
        elif "Sq. Yards" in x:
            pattern = r'(\d+(\.\d+)?)'
            match = re.search(pattern, x)
            if match:
                number = float(match.group(0))
                return round(number * 9)
        
        elif "Acres" in x:
            pattern = r'(\d+(\.\d+)?)'
            match = re.search(pattern, x)
            if match:
                number = float(match.group(0))
                return round(number * 43560)

        elif "Cents" in x:
            pattern = r'(\d+(\.\d+)?)'
            match = re.search(pattern, x)
            if match:
                number = float(match.group(0))
                return round(number * 435.6)
    
        elif "Guntha" in x:
            pattern = r'(\d+(\.\d+)?)'
            match = re.search(pattern, x)
            if match:
                number = float(match.group(0))
                return round(number * 1089)
        
        elif "Grounds" in x:
            pattern = r'(\d+(\.\d+)?)'
            match = re.search(pattern, x)
            if match:
                number = float(match.group(0))
                return round(number * 2400)
        
    except:
        return None
 
```

```python
temp_df.loc[:,"total_sqft"] = temp_df["total_sqft"].apply(convert_sqft_to_num)
```

Removing the outliers in `price_per_sqft` feature based on mean and one standard deviation for each group

```python
def remove_price_per_sqft_outliers(df):
    df_out = pd.DataFrame()
    for key, sub_df in df.groupby('location'):
        mean = np.mean(sub_df.price_per_sqft)
        std = np.std(sub_df.price_per_sqft)
        
        reduced_df = sub_df[(sub_df.price_per_sqft > (mean-std)) & (sub_df.price_per_sqft <= (mean+std))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
```

##### Let's check if for a given location how does the 2 BHK and 3 BHK property prices look like

```python
def plot_scatter_chart(df,location):
    bhk2 = df[(df.location==location) & (df.bhk==2)]
    bhk3 = df[(df.location==location) & (df.bhk==3)]
    plt.rcParams['figure.figsize'] = (12,6)
    plt.scatter(bhk2.total_sqft,bhk2.price,color='blue',label='2 BHK', s=50)
    plt.scatter(bhk3.total_sqft,bhk3.price,marker='+', color='green',label='3 BHK', s=50)
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price (Lakh Indian Rupees)")
    plt.title(location)
    plt.legend()
plot_scatter_chart(df4,"Rajaji Nagar")
```

```python
plot_scatter_chart(df4,"Hebbal")
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
df5 = remove_bhk_outliers(df4)
df5.shape
```
